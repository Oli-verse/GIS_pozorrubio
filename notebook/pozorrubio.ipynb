'''
This code creates an interactive map of Pozorrubio, Philippines, using the Folium library. It loads barangay coordinates
from a CSV file, adds markers for each barangay, and overlays a GeoJSON layer to show the municipality boundary. 
A custom legend is also included to explain the markers on the map. Finally, the map is saved as an HTML file for viewing in a web browser.

AI agents used:
Google Gemini - Research and Implementation
Claude - Minimal code draft and debugging
Copilot - Code completion
'''

```python
import folium
import pandas as pd
import branca
```

```python
#load the barangay coordinates from the CSV file
barangays = pd.read_csv('data/barangay_coords.csv')
universities = pd.read_csv('data/university_coords.csv')
barangays.head()
```

```python
#Generate the map centered on Pozorrubio
map = folium.Map(location=[16.11, 120.54], zoom_start=13,tiles='CartoDB positron')
map
```

```python
#Markers for each barangay
for _, row in barangays.iterrows():
    folium.Marker(
        location=[row['origin_latitude'], row['origin_longitude']],
        tooltip=row['origin_barangay'],
        fill_color='blue',
        fill_opacity=0.7,
        popup=folium.Popup(f"<b>Barangay:</b> {row['origin_barangay']}", max_width=200)
    ).add_to(map)

map
```

```python
for _, row in universities.iterrows():
    folium.Marker(
        location=[row["uni_latitude"], row["uni_longitude"]],
        tooltip=row["university_name"],
        popup=folium.Popup(f"<b>University:</b> {row['university_name']}", max_width=200),
        icon=folium.Icon(color="red", icon="graduation-cap", prefix="fa")
    ).add_to(map)

map
```

```python
# Add the GeoJSON layer for the municipality boundary
folium.GeoJson ("data/bgysubmuns-municity-105530000.0.1.json",
               name='pozorubio_boundary',
               style_function=lambda x: {
                    "color": "black",
                    "weight": 2,
                    "fillColor": "yellow",
                    "fillOpacity": 0.1
                }
               ).add_to(map)
map
```

```python
# Create a custom legend
legend_html = '''
{% macro html(this, kwargs) %}
<div style="position: fixed; 
     bottom: 50px; left: 50px; width: 200px; height: 65px; 
     border:2px solid grey; z-index:9999; font-size:14px;
     background-color:white; opacity: 0.85;">
     &nbsp; <b>Legend</b> <br>
     &nbsp; Barangay Name &nbsp; <i class="fa fa-circle" style="color:blue"></i><br>
     &nbsp; University Name &nbsp; <i class="fa fa-graduation-cap" style="color:red"></i><br>
</div>
{% endmacro %}
'''

legend = branca.element.MacroElement()
legend._template = branca.element.Template(legend_html)

# Add the legend to the map
map.get_root().add_child(legend)
map
```

```python
# Save the map to an HTML file
map.save('output/pozorrubio_map.html')
print("Map has been saved to 'output/pozorrubio_map.html'")
map
```
